```python
# generator to print even numbers
def print_even(test_list):
    for i in test_list:
        if i % 2 == 0:
            yield i
```

```python
test_list = [1, 4, 5, 6, 7]
```

```python
# printing initial list
print("The original list is : " + str(test_list))
 
# printing even numbers
print("The even numbers in list are : ", end=" ")
for j in print_even(test_list):
    print(j, end=" ")
```

```python
def even_numbers(limit):
    for i in range(0, limit, 2):
        yield i
    # return
```

Notice that the even_numbers() function doesn't return a list or tuple, but instead generates the sequence of numbers on the fly. This can be useful when working with large datasets or infinite sequences, as it avoids the need to store the entire sequence in memory at once.

Overall, generators and the yield keyword provide a powerful and flexible way to create iterators in Python.

### Example 1: Fibonacci sequence
The Fibonacci sequence is a famous sequence of numbers where each number is the sum of the two preceding ones, starting from 0 and 1. Here's an example of a generator function that generates the Fibonacci sequence up to a given limit:

```python
def fibonacci(limit):
    a, b = 0, 1
    
    while a < limit:
        yield a
        a, b = b, a + b
```

```python
# Example usage
for num in fibonacci(10):
    print(num)
```

Notice that the fibonacci() function doesn't generate the entire sequence at once, but instead generates each number on the fly as it is needed.



### Example 2: Squares of numbers
Here's an example of a generator function that generates the squares of numbers up to a given limit:

```python
def square_numbers(limit):
    for i in range(limit):
        yield i**2   # 'i' raised to 2
```

```python
# Example usage
for num in square_numbers(5):
    print(num)
```

```python
# Using generator 
def PowTwoGen(max=0):
    n = 0
    while n < max:
        yield 2 ** n
        n += 1



# Create a generator object that generates powers of two up to 10
my_gen = PowTwoGen(10)

# Iterate over the generator object and print each value
for num in my_gen:
    print(num)
```

```python
def fibonacci_numbers(nums):
    x, y = 0, 1
    for _ in range(nums):
        x, y = y, x+y
        yield x

def square(nums):
    for num in nums:
        yield num**2

print(sum(square(fibonacci_numbers(10))))

# Output: 4895
```
