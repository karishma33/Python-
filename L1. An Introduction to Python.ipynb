The `>>=` operator is a compound assignment operator in Python that performs a right shift operation on a variable and assigns the result back to the variable. The `>>` operator shifts the bits of a number to the right by a specified number of positions, with the empty bits on the left being filled with zeroes.


````python
x = 10  # binary representation is 1010
x >>= 1  # perform a right shift by 1 bit- after shifting 0101
print(x)  # output: 5
````

````python
x = 10  # binary representation is 1010
x <<= 3  # perform a left shift by 3 bits
print(x)  # output: 80


# In this example, the initial value of x 
# is 10 which has a binary representation 
# of 1010. The <<= operator shifts this 
# value by 3 bits to the left, resulting in the 
# binary value 1010000, which is equal to the decimal 
# value 80. Finally, the result 80 
# is assigned back to the variable x.


# In short, x <<= 3 is equivalent to x = x << 3.
````

### Identity Operator

````python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)  # Output: True, because a and b refer to the same object in memory
print(a is c)  # Output: False, because a and c refer to different objects in memory
````

````python
print(id(a))
print(id(b))
````

````python
print(id(c))
````

````python
print(a is not b)
````

In Python, the id() function returns the unique identifier (memory address) of an object. It is an integer value that is guaranteed to be unique and constant for the lifetime of the object.

In the given code snippet, a is a list object containing the elements [1, 2, 3]. When we assign a to b using the assignment operator =, b refers to the same object as a, so the id() function will return the same value for both a and b.

On the other hand, when we create a new list object containing the same elements [1, 2, 3] and assign it to c, c refers to a different object than a and b, so the id() function will return a different value for c.

So, in the given code snippet, the id(a) and id(b) will return the same value (the memory address of the list object [1, 2, 3]), while id(c) will return a different value (the memory address of a new list object containing the elements [1, 2, 3]).

````python
# The is not operator returns True if two variables
# do not refer to the same object in memory, and
# False otherwise. Here's an example:

````

````python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is not b)  # Output: False, because a and b refer to the same object in memory
print(a is not c)  # Output: True, because a and c refer to different objects in memory
````

## Collection Data-Types

So far we have seen built-in types like: *int, float* and *bool*. int, float, and bool are considered to be simple or primitive data types because their values are not composed of any smaller parts. **They cannot be broken down.**

On the other hand, *strings, lists, sets, dictionaries* and *tuples* are different from the others because they are made up of smaller pieces. In the case of strings, they are made up of smaller strings each containing one character.

**Types that are comprised of smaller pieces are called collection data types.** Depending on what we are doing, we may want to treat a collection data type as a single entity (the whole), or we may want to access its parts. This ambiguity is useful.

### Strings

Strings are a collection of characters. They can store a line of text. Strings are arrays of bytes representing Unicode characters.

Interestingly enough, Python does not have a character datatype. It stores characters as a string of length 1.

## The if statement

This is used to decide if a statement / a group of statements should be executed, based on a condition.

````python
a = 5
b = 6

if a<b:
      
    print(a)
    print("Yes!")
````

Observe a few things:
- The syntax is 
```python
if <expression>:
    <statements>
```
- The statements after 'if' are executed if the expression evaluates to True.
- Notice that we do not have a bracket to denote the part of code which is to be executed if the condition is true.
- Instead, we use indentation to denote that block of statements. A constant indent tells Python that those statements belong to one code block.

For a complicated condition(using logical operators and/or expressions), we enclose all of it in a bracket.

Now, we can also extend this to execute another block of statements if the condition evaluates to False.

````python
a = 5
b = 6
if a>b:
    print(a)
    print("a is greater than b!")
else:
    print(b)
    print("a is less than b!")
````

Thus, the keywords 'if' and 'else' are used to execute statements based on a condition.

What do you do if you want to have multiple condition based statement execution?

For example, if a person is considered:
- Short if his height is less than 4 feet.
- Average if his height is between 4 and 5 feet.
- Tall if his height is more than 5 feet.

We need another keyword here to make our lives easier, called 'elif'.

````python
height = float(input("Enter your height: "))
if height<4:
    print("Short")
elif height<5:
    print("Average")
else:
    print("Tall")
````

The importance of indentation is paramount in Python. It replaces the cumbersome curly brackets that exist in C, and which if you forget to close, can lead to so many problems.

## continue and break

The continue statement is used in loops to skip all the following statements in the current code block, and proceed with the next iteration.

````python
a = 8
while(a>0):
    a=a-1
    if(a<5 and a>1):
        continue
    print(a)
````

````python
a = 10
while(a>0):
    a=a-1
    print("Hello")
    if(a<5 and a>1):
        continue
    print(a)
````

The break statement exits the loop being executed.

#### Pass 

In Python programming, the pass statement is a null statement which can be used as a placeholder for future code.

Suppose we have a loop or a function that is not implemented yet, but we want to implement it in the future. In such cases, we can use the pass statement.

````python
# Using pass With Conditional Statement
n = 10
# use pass inside if statement
if n > 10:
    pass
    

print('Hello')
````

### assert

````python
def divide(a, b):
    assert b != 0, "Cannot divide by zero"
    return a / b
print(divide(10, 2))  # Output: 5.0
````

````python
print(divide(5, 0))  # Raises an AssertionError: Cannot divide by zero
````

````python
divide(4,2)
divide(10,2)
````

````python
print(divide(4,2))
print(divide(10,2))
````

````python
def calculate_area(length, width):
    assert length > 0 and width > 0, "Length and width must be greater than 0."
    area = length * width
    return area
````

````python
calculate_area(5,6)
````

````python
calculate_area(5,0)
````

````python
def calculate_discount(price, discount):
    assert 0 <= discount <= 1, "Discount must be between 0 and 1"
    discounted_price = price * (1 - discount)
    return discounted_price

# Example usage:
print(calculate_discount(100, 0.2))  # Output: 80.0
print(calculate_discount(50, 1.5))  # Raises an AssertionError: Discount must be between 0 and 1
````
