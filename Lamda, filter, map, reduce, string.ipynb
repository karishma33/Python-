The output of the code includes boolean values because the `isinstance()` function is checking if each element in `list1` is an instance of the `int` type. However, in Python, `bool` is a subclass of `int`, so boolean values (`True` and `False`) are also considered instances of `int`.

This behavior stems from the fact that in Python, `True` is equivalent to `1` and `False` is equivalent to `0`. Therefore, when the `isinstance()` function checks if an element is an instance of `int`, it returns `True` for boolean values as well.

If you want to exclude boolean values from the filtered list, you can modify the code to explicitly exclude them:

```python

print(list1_only_int)
```

This will remove boolean values from the filtered list, and you will only have the integer values in the output:

```
[12, -21, -45]
```

# reduce

The functools module is for higher-order functions. Functions that act on or return other functions. In general, any callable object can be treated as a function for the purposes of this module.

The functools module provides the following function functools.reduce()

functools.reduce()

Apply function of two arguments cumulatively to the items of iterable, from left to right, so as to reduce the iterable to a single value. -python docs

````python
import math 
math.ceil(4.5)
````

````python
# python code to demonstrate working of reduce()
 
# importing functools for reduce()
import functools as ft
 
# initializing list
list_1 = [1,2,3,4,5]
# using reduce to compute sum of list
print("The sum of the list elements is: ", end="")
print(ft.reduce(lambda x, y: x+y, list_1))
````

### Binary functions 

A binary function is a function that takes two arguments and produces a result. In other words, it is a function that operates on a pair of elements and returns a single value. The term "binary" in this context refers to the fact that the function has arity 2, meaning it takes two inputs.

Binary functions are commonly used in mathematics and computer science. They can perform various operations such as addition, subtraction, multiplication, division, comparison, logical operations, and more. Examples of binary functions include:

- Addition: `f(x, y) = x + y`
- Subtraction: `f(x, y) = x - y`
- Multiplication: `f(x, y) = x * y`
- Division: `f(x, y) = x / y`
- Maximum: `f(x, y) = max(x, y)`
- Minimum: `f(x, y) = min(x, y)`
- Logical AND: `f(x, y) = x AND y`
- Logical OR: `f(x, y) = x OR y`

In programming, binary functions are often implemented as methods or operators that take two arguments and return a result. For example, the `+` operator in Python is a binary function for addition, and the `>` operator is a binary function for comparison.

Overall, binary functions provide a way to combine two inputs and produce a meaningful output based on a specific operation or logic.

````python
# Example 2: If the function parameter is not mentioned, 
# by default it will perform an addition operation

# It will return an iterator that yields all intermediate values. We can convert to a list by using list() constructor.

import itertools
import operator


num1=itertools.accumulate([1,2,3,4,5])
print (num1)


#Converting iterator to list object.
print (list(num1))#Output:[1, 3, 6, 10, 15]
#using reduce() for same function
from functools import reduce
r1=reduce(operator.add,[1,2,3,4,5])
print (r1)#Output:15
````

````python
# Example 3:Function argument is given as max(), to find a running maximum

from itertools import accumulate
num4=accumulate([2,4,6,3,1],max)
print (list(num4))#Output:[2, 4, 6, 6, 6]
````

````python
# Example 4: If the initial value is mentioned, it will start accumulating from the initial value.

from itertools import accumulate
import operator
  
#If initial parameter is mentioned, it will start accumulating from the initial value.
#It will contain more than one element in the ouptut iterable.
num2=accumulate([1,2,3,4,5],operator.add,initial=10)
print (list(num2))#Output:[10, 11, 13, 16, 20, 25]
````

````python
# Example 6: If iterable contains one element and the initial parameter is not mentioned, it will return that element.

from itertools import accumulate

num2=accumulate([5],lambda x,y:x+y)
print (list(num2))#Output:[5]
````

````python
# Example 7: Iterating through the iterator using for loop

# Return type is an iterator. We can iterate through iterator using for loop also.

from itertools import accumulate
num2=accumulate([5,6,7],lambda x,y:x+y)
for i in num2:
    print (i)
````

Conclusion:

reduce() function is supported by the functools module.

accumulate() function is supported by the itertools module.

reduce() will return only an accumulated value.

accumulate() will return the running accumulated value. Like we can find running max, running product, running sum of an iterable using the accumulate() function.

accumulate() returns an iterator.
